```python
import gdown
gdown.download('https://bit.ly/3pK7iuu','ns_book7.csv',quiet=False)

import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv')
ns_book7.head()

import matplotlib.pyplot as plt
plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
plt.show()

# figure 객체 이용

# 그래프 크기 바꾸기 : figsize 매개변수
plt.figure(figsize=(9,6)) # 가로 9인치, 높이 6인치인 피겨 객체 생성(기본은 6,4임)
plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
plt.show()
print(plt.rcParams['figure.figsize']) # 프로그램마다 기본 사이즈가 다른 경우가 있음
print(plt.rcParams['figure.dpi']) # dot per inch을 출력함. figsize로 설정한 인치가 아닌경우도 있는데 해상도 때문임
plt.figure(figsize=(900/100,600/100)) # 900x600픽셀인데 맷플롯립 dpi 기본값이 72(이 프로그램은 100)이므로 나눠준다.
plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
plt.show()

%config InlineBackend.print_figure_kwargs = {'bbox_inches':None} # 코랩의 타이트 레이아웃 미설정(주변에 공백 많이 생김)

plt.figure(figsize=(900/72,600/72))
plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
plt.show()

%config InlineBackend.print_figure_kwargs = {'bbox_inches':'tight'} # 타이트 레이아웃 다시 돌리기

# 그래프 크기 바꾸기 : dpi 매개변수
plt.figure(dpi=144) # dpi 매개변수를 늘리면 인치당 픽셀 수가 늘어나므로 그래프도 커진다. 동그라미(마커)의 크기도 커졌다.
plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
plt.show()

# rcParams 객체 : 맷플롯립 그래프의 기본값을 관리한다.
plt.rcParams['figure.dpi'] = 100 # dpi 기본값을 변경해 해상도를 높인다.
plt.rcParams['scatter.marker'] = '*' # 산점도 마커 모양을 별로 바꾼다.
plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
plt.show()

plt.scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1,marker='+') # marker 매개변수를 이용해 바꿀 수도 있다.
plt.show()

# 여러 개의 서브플롯 출력하기 : Axes 클래스의 객체를 이용, 하나의 서브플롯은 두 개 이상의 축을 포함한다.
# x축,y축에는 틱(눈금), 축의 이름인 레이블,선,마커,등이 있다.
# 서브플롯 그리기 : subplots()
fig, axs = plt.subplots(2)
axs[0].scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1) # 첫번째 그래프
axs[1].hist(ns_book7['대출건수'],bins=100) # 두번째 그래프
axs[1].set_yscale(value='log')
fig.show() # 그래프가 두개여서 각각의 사이즈가 작다.

fig, axs = plt.subplots(2,figsize=(6,8)) # figsize 매개변수 지정이 가능하다.
axs[0].scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
axs[0].set_title('scatter plot') # 서브 플롯의 제목
axs[1].hist(ns_book7['대출건수'],bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale(value='log')
fig.show()

fig, axs = plt.subplots(1,2,figsize=(10,4)) # subplot(2,3) = 행이 2개 열이 3
axs[0].scatter(ns_book7['도서권수'],ns_book7['대출건수'],alpha=0.1)
axs[0].set_title('scatter plot')
axs[0].set_xlabel('number of books')
axs[0].set_ylabel('borrow counts')

axs[1].hist(ns_book7['대출건수'],bins=100)
axs[0].set_title('histogram')
axs[0].set_xlabel('borrow counts')
axs[0].set_ylabel('frequency')
axs[1].set_yscale('log')

fig.show()
```
