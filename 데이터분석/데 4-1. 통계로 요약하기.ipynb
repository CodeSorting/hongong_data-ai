```python
import gdown
gdown.download('https://bit.ly/3736JW1','ns_book6.csv',quiet=False)
import pandas as pd
ns_book6 = pd.read_csv('ns_book6.csv',low_memory=False)
ns_book6.head()

ns_book6.describe() # 몇가지 기술통계에 대해 기술 count,mean,std,min,50%(중앙값)
# sum(ns_book6['도서권수']==0) # 3206개나 있으나 1%도 안되기 때문에 얘를 제외하고 보자.
ns_book7 = ns_book6[ns_book6['도서권수']>0]
ns_book7.describe(percentiles=[0.3,0.6,0.9]) # percentiles로 25% 50% 75%가 아닌 30% 60% 90%가 나온다
ns_book7.describe(include='object') # 수치가 아닌 다른 데이터 타입의 열의 기술통계, unique(고유한 값의 개수),top(가장 많이 등장하는 값),freq(top행에 등장하는 항목의 빈도수)가 나온다.
ns_book7['대출건수'].mean() # 대출건수 열의 평균
ns_book7['대출건수'].median() # 대출건수 열의 중앙값
ns_book7['대출건수'].mode() # 대출건수 열의 최빈값
ns_book7['대출건수'].drop_duplicates().median() # 중복값 제거하고 중앙값 구하기
ns_book7['대출건수'].max() # 대출건수 열의 최대값
ns_book7['대출건수'].min() # 대출건수 열의 최소값
ns_book7['대출건수'].std() # 대출건수 열의 표준편차
ns_book7['대출건수'].var() # 대출건수 열의 분산
ns_book7['대출건수'].quantile(0.25) # 대출건수 분위수 값 계산(하위 25%)
ns_book7['대출건수'].quantile([0.25,0.5,0.75])
pd.Series([1,2,3,4,5]).quantile(0.9) # 4.6이 나옴(4와 5 사이에서 90%에 해당하는 분위수를 찾을 때, 4의 분위수 0.75와 5의 분위수 1.0을 사용해 비례식으로 계산한다.)
borrow_10_flag = ns_book7['대출건수']<10 # 대출건수가 10보다 작은 백분위 구하기 위한 불리언 배열
borrow_10_flag.mean()

# 넘파이의 기술통계 함수
import numpy as np
np.mean(ns_book7['대출건수'])
np.average(ns_book7['대출건수'], weights=1/ns_book7['대출건수']) # weights 매개변수로 가중치를 설정해 평균을 구한다.
np.mean(ns_book7['대출건수']/ns_book7['도서권수'])
np.median(ns_book7['대출건수'])
np.min(ns_book7['대출건수'])
np.max(ns_book7['대출건수'])
np.quantile(ns_book7['대출건수'],[0.25,0.5,0.75])
np.var(ns_book7['대출건수']) # n으로 나눔
ns_book7['대출건수'].var() # 위의 분산과 다름. 식을 구할 때 얘는 n-1로 나누기 때문이다.
np.std(ns_book7['대출건수'])
# 최빈값 구하기
values, counts = np.unique(ns_book7['도서명'], return_counts=True)
max_idx = np.argmax(counts)
values[max_idx]
```
