```python
import gdown
gdown.download('https://bit.ly/3eecMKZ','남산도서관 장서 대출목록 (2021년 04월).csv', quiet=False)

import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv', 'rb') as f: # 파일 열고 f라고 저장, rb=바이너리 읽기모드
  d = f.readline()
print(chardet.detect(d)) # 이런 인코딩 사용 (EUC-KR 사용하므로 UTF-8을 사용하거나 EUC-KR을 인코딩으로 정해야함)

with open('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR') as f: # EUC-KR으로 인코딩
  print(f.readline()) # 한 줄씩 출력
  print(f.readline())

import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR',low_memory=False)
# 판다스가 파일을 읽을 때 타입을 자동으로 정하고 조금씩 파일을 나눠서 읽는다.
# 이때 실제데이터가 자동으로 바뀐 타입과 맞지 않을 때 에러가 발생한다.
# ISBN,세트 ISBN, 주제분류번호를 dtype을 이용하거나 low_memory=False를 해줘서 한번에 읽게 한다.
# df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR',dtype={'ISBN' : str,'세트 ISBN' : str, '주제분류번호' : str})
df.head() # head()로 처음 5개 파악

df.to_csv('ns_202104.csv') # 데이터 프레임을 csv파일로 변환
with open('ns_202104.csv') as f:
  for i in range(3):
    print(f.readline(),end='') # end=''를 하면 줄바꿈이 한번 더 안된다.

ns_df = pd.read_csv('ns_202104.csv',low_memory=False)
print(ns_df.head())

ns_df = pd.read_csv('ns_202104.csv',index_col=0,low_memory=False) # 행 인덱스가 이미 있다는 걸 알려줌(unnamed:0이라는 추가 열이 안 생김)
print(ns_df.head())

# 아니면 애초에 위에서 변환할 때 index=False로 놓는다.
df.to_csv('ns_202104.csv',index=False)
print(df.head())
```
