```python
import gdown
gdown.download('https://bit.ly/3pK7iuu','ns_book7.csv',quiet=False)

import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv')

from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(ns_book7, random_state=42)
print(len(train_set),len(test_set))

x_train = train_set[['도서권수']] # 2차원 배열 (282577,1)
y_train = train_set['대출건수'] # 1차원 배열 (282577,)
# 2차원 배열 = 행방향이 샘플, 열방향이 속성

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)

# score()로 평가, 결정
x_test = test_set[['도서권수']]
y_test = test_set['대출건수']
lr.score(x_test,y_test) # 0.1... 점수가 별로 안 좋음. 도서권수로 대출건수를 평가하는 것은 상관관계가 역시 거의 없음

lr.fit(y_train.to_frame(),y_train) # 대출건수로 대출건수를 에측하면 무조건 1 나옴 ㅋㅋㅋ , y_train이 1차원배열이므로 2차원으로 바꿔준다.
lr.score(y_test.to_frame(),y_test)

# 선형 회귀 : y = ax + b coef, intercept
print(lr.coef_,lr.intercept_) # 기울기, y절편

# 로지스틱 회귀 : 그 중에서도 이중 분류를 사용하며 양성 클래스, 음성 클래스로 나누기 위해 평균보다 크면 양성, 작으면 음성으로 한다.
borrow_mean = ns_book7['대출건수'].mean()
y_train_c = y_train > borrow_mean
y_test_c = y_test > borrow_mean

from sklearn.linear_model import LogisticRegression

logr = LogisticRegression()
logr.fit(x_train,y_train_c)
logr.score(x_test,y_test_c) # 0.71.. 정답 맞춘 비율

# 양성 클래스 비율보다 음성 클래스 비율이 더 많음!
y_test_c.value_counts() # 양성 31% 음성 69%

from sklearn.dummy import DummyClassifier

dc = DummyClassifier()
dc.fit(x_train,y_train_c)
dc.score(x_test,y_test_c) # 0.5.. 정답 맞춘 비율
```
