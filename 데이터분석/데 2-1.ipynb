```python
import json
d = {"name": "혼자 공부하는 데이터 분석"}
print(d['name'])
print(json.dumps(d)) # json.dumps = 파이썬 -> json
print(json.loads(json.dumps(d))) # json.loads = json -> 파이썬
print(type(json.loads(json.dumps(d))))

d3 = json.loads('{"name": "혼자 공부하는 데이터 분석", "author": "박해선", "year": 2022}')
print(d3['name'])
print(d3['author'])
print(d3['year'])

d4_str = """
[
  {"name": "혼자 공부하는 데이터 분석", "author": "박해선", "year": 2022},
  {"name": "혼자 공부하는 머신러닝", "author": "박해선", "year": 2020}
]
"""
d4 = json.loads(d4_str) # json -> 파이썬
print(d4[0]['name']) # 혼자 공부하는 데이터 분석

import pandas as pd
pd.read_json(d4_str) # json -> 데이터프레임
pd.DataFrame(d4) # 파이썬 -> 데이터프레임

#xml 다루기
x_str = """
<root>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</root>
"""
import xml.etree.ElementTree as et # text,tag,attrib 속성
root = et.fromstring(x_str) # xml -> Element 객체
print(root.tag) # root
print(root.attrib) # {}
print(root[0].tag) # name
print(root[0].text) # 혼자 공부하는 데이터 분석
print(root[1].tag) # author
print(root[1].text) # 박해선
print(root[2].tag) # year
print(root[2].text) # 2022

book_childs = list(root) # Element -> 리스트
print(book_childs) # [<Element 'name' at 0x7e53f8e1b4c0>, <Element 'author' at 0x7e53f8552b10>, <Element 'year' at 0x7e53f85528e0>]

name,author,year = book_childs # book_childs에서 순서대로 찾는것은 위험하다. 순서가 항상 일치하지는 않기 때문이다.
print(name.text) # 혼자 공부하는 데이터 분석
print(author.text) # 박해선
print(year.text) # 2022

name = root.findtext('name') #findtext()
author = root.findtext('author')
year = root.findtext('year')
print(name) # 혼자 공부하는 데이터 분석
print(author) # 박해선
print(year) # 2022
print()

x2_str = """
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>
  <book>
    <name>혼자 공부하는 머신러닝</name>
    <author>박해선</author>
    <year>2020</year>
  </book>
</books>
"""
books = et.fromstring(x2_str) # xml -> Element 객체
for book in books.findall('book'): # findall()로 자식 엘리먼트 <book> 반환
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')
  print(name) # 혼자 공부하는 데이터 분석
  print(author) # 박해선
  print(year) # 2022
  print()
```
