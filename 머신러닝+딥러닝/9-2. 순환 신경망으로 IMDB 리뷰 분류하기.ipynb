```python
from tensorflow.keras.datasets import imdb
(train_input,train_target), (test_input,test_target) = imdb.load_data(num_words=300) #   300   .
print(train_input.shape, test_input.shape)
print(train_input[0]) # 218   . 2   , 1  , 0 .
print(train_target[:10])

from sklearn.model_selection import train_test_split
train_input, val_input, train_target, val_target = train_test_split(train_input, train_target, test_size=0.2, random_state=42) #  5000

import numpy as np
lengths = np.array([len(x) for x in train_input])
print(np.mean(lengths), np.median(lengths)) # 239, 178
import matplotlib.pyplot as plt
plt.hist(lengths)
plt.xlabel('length')
plt.ylabel('frequency')
plt.show() #   . 100   1000      .
# 100     100    (0 ).

from tensorflow.keras.preprocessing.sequence import pad_sequences # .
train_seq = pad_sequences(train_input, maxlen=100) # 100  . , maxlen   .    .     truncating='post' .  pre.
print(train_seq.shape) # (20000,100)
print(train_seq[0])
print(train_seq[5]) # 100   0 .
val_seq = pad_sequences(val_input, maxlen=100)
print(val_seq.shape) # (5000,100)

#   
from tensorflow import keras
model = keras.Sequential()
model.add(keras.layers.SimpleRNN(8, input_shape=(100,300))) # , input_shape( ,               (300  ) )
# SimpleRNN    tanh .
model.add(keras.layers.Dense(1, activation='sigmoid')) #     

train_oh = keras.utils.to_categorical(train_seq) #     (   (20000,100) (20000,100,300) )
print(train_oh.shape)
print(train_oh[0][0][:12])
val_oh = keras.utils.to_categorical(val_seq)
model.summary() # 300x8(      8 ) + 8x8(        .) + 8()= 2472

#   
rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4)
model.compile(optimizer=rmsprop, loss='binary_crossentropy', metrics=['accuracy'])
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-simplernn-model.h5')
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
history = model.fit(train_oh, train_target, epochs=100, batch_size=64, validation_data=(val_oh, val_target), callbacks=[checkpoint_cb, early_stopping_cb])

plt.plot(history.history['loss']) #  
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
print(train_seq.nbytes, train_oh.nbytes) # nbytes     : 8000000 2400000000 ( .)

#    :       .
model2 = keras.Sequential()
model2.add(keras.layers.Embedding(300, 16, input_length=100)) #   (300    ),  ,   (100)
model2.add(keras.layers.SimpleRNN(8))
model2.add(keras.layers.Dense(1, activation='sigmoid'))
model2.summary()

rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4)
model2.compile(optimizer=rmsprop, loss='binary_crossentropy', metrics=['accuracy'])
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-simplernn-model.h5')
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
history = model2.fit(train_seq, train_target, epochs=100, batch_size=64, validation_data=(val_seq, val_target), callbacks=[checkpoint_cb, early_stopping_cb]) #  validation_data , batch_size  ,   .  CPU 

plt.plot(history.history['loss']) #  
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```
