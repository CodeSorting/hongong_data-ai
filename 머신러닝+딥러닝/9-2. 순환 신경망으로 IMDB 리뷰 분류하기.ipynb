```python
from tensorflow.keras.datasets import imdb
(train_input,train_target), (test_input,test_target) = imdb.load_data(num_words=300) # 최대 글자수 300개 그 이상은 짜른다.
print(train_input.shape, test_input.shape)
print(train_input[0]) # 218개의 단어를 정수로 변경함. 2는 사전에 없는 토큰, 1은 문장의 시작, 0은 패딩이다.
print(train_target[:10])

from sklearn.model_selection import train_test_split
train_input, val_input, train_target, val_target = train_test_split(train_input, train_target, test_size=0.2, random_state=42) # 검증세트 5000개

import numpy as np
lengths = np.array([len(x) for x in train_input])
print(np.mean(lengths), np.median(lengths)) # 239, 178
import matplotlib.pyplot as plt
plt.hist(lengths)
plt.xlabel('length')
plt.ylabel('frequency')
plt.show() # 길이가 한쪽으로 치우쳐져있다. 100자 이하인게 대부분이고 1000자 이상인 것도 있어서 평균이 ㅈ같이 나온것이다.
# 100자 이하인 작은 리뷰는 길이를 100으로 맞추기 위해 패딩을 사용(0으로 채움)한다.

from tensorflow.keras.preprocessing.sequence import pad_sequences # 패딩해준다.
train_seq = pad_sequences(train_input, maxlen=100) # 100자 까지 패딩한다. 단, maxlen보다 길다면 앞부분을 자른다. 뒷정보가 유용하다고 판단하기 때문이다. 만약 뒷정보를 자르고 싶다면 truncating='post'라고 하면된다. 기본은 pre이다.
print(train_seq.shape) # (20000,100)
print(train_seq[0])
print(train_seq[5]) # 100보다 작았었음 앞을 0으로 패딩함.
val_seq = pad_sequences(val_input, maxlen=100)
print(val_seq.shape) # (5000,100)

# 순환 신경망 만들기
from tensorflow import keras
model = keras.Sequential()
model.add(keras.layers.SimpleRNN(8, input_shape=(100,300))) # 뉴런개수, input_shape(샘플의 길이, 단어의 고유값이 크더라도 활성화 출력이 크지 않기 위해 원 핫 인코딩으로 표현하기 위한 배열의 길이(300개의 단어만 사용) )
# SimpleRNN의 기본 활성화 함수는 tanh를 사용한다.
model.add(keras.layers.Dense(1, activation='sigmoid')) # 부정인지 긍정인지 분류하는 이진분류를 사용할것

train_oh = keras.utils.to_categorical(train_seq) # 원 핫 인코딩을 위한 유틸리티(원 핫 인코딩으로 (20000,100)을 (20000,100,300)으로 변경)
print(train_oh.shape)
print(train_oh[0][0][:12])
val_oh = keras.utils.to_categorical(val_seq)
model.summary() # 300x8(원 핫 인코딩 배열이 순환층의 뉴런 8개와 완전연결) + 8x8(순환층의 은닉상태가 다음스텝에 사용되기 위해 또 다른 가중치에 곱해짐.) + 8(절편)= 2472

# 순환 신경망 훈련하기
rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4)
model.compile(optimizer=rmsprop, loss='binary_crossentropy', metrics=['accuracy'])
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-simplernn-model.h5')
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
history = model.fit(train_oh, train_target, epochs=100, batch_size=64, validation_data=(val_oh, val_target), callbacks=[checkpoint_cb, early_stopping_cb])

plt.plot(history.history['loss']) # 그래프 그리기
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
print(train_seq.nbytes, train_oh.nbytes) # nbytes 속성으로 데이터 크기 확인 : 8000000 2400000000 (존나 크다.)

# 단어 임베딩 사용하기 : 각 단어를 고정된 크기의 실수 벡터로 바꾸어준다.
model2 = keras.Sequential()
model2.add(keras.layers.Embedding(300, 16, input_length=100)) # 어휘 사전의 크기(300개 단어만 사용하기로 했기 때문),임베딩 벡터의 크기, 입력 시퀸스의 길이(100)
model2.add(keras.layers.SimpleRNN(8))
model2.add(keras.layers.Dense(1, activation='sigmoid'))
model2.summary()

rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4)
model2.compile(optimizer=rmsprop, loss='binary_crossentropy', metrics=['accuracy'])
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-simplernn-model.h5')
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
history = model2.fit(train_seq, train_target, epochs=100, batch_size=64, validation_data=(val_seq, val_target), callbacks=[checkpoint_cb, early_stopping_cb]) # 검증용으로 validation_data 사용, batch_size로 한번에 처리,계산할 샘플 개수를 말해줌. 다중 CPU 설정

plt.plot(history.history['loss']) # 그래프 그리기
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```
