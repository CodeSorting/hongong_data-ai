```python
from tensorflow import keras

# keras.layers.Conv2D(10, kernel_size=(3,3),activation='relu') # 필터 개수, 커널의 크기, 활성화함수
# 특성 맵의 크기를 동일하게 만들기 위해 입력 주위에 0을 패딩하는 것을 세임 패딩이라고 함.
# 패딩 없이 순수한 입력 배열에서만 합성곱을 하는 것을 밸리드 패딩이라고 함.
# keras.layers.Conv2D(10, kernel_size=(3,3),activation='relu',padding='same') # padding 매개변수를 same으로 하면 세임패딩이 된다.

# 스트라이드 : 도장(필터)을 이동하는 칸의 개수, 기본값은 1
# keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same', strides=1) (x,y)처럼 strides는 오른쪽, 아래쪽으로 이동하는 크기를 튜플을 통해 지정할 수 있다.

# 풀링 : 특성맵의 개수는 유지하고 가로,세로 크기를 줄인다.
# 가로 세로 크기를 줄이기 위해 특정 크기의 커널로 풀링을 한다. 하지만 가중치는 없어서 가장 큰 값(최대풀링)이나 평균값(평균 풀링)을 채택한다. 겹치지 않고 이동한다.
# keras.layers.MaxPooling2D(2, strides=2, padding='valid') 풀링의 크기(2는 가로 세로 절반 줄인다는 뜻, (2,3)처럼도 가능), strides=풀링의 크기라 따로 안 정해도 된다,padding='valid'는 더이상 패딩을 안한다는 뜻이다.
# 평균 풀링은 AveragePooling2D이다.

"""
입력이 (4,4)이고 필터가 (3,3)일 때
순서
1. 세임 패딩으로 (4,4,3) 특성맵 3개 만들기
2. (2,2)로 최대 풀링 해보기 -> (2,2,3)
3. 1차원으로 펼치기(2,2,3) -> (12) (Flatten 함수 사용)
4. 밀집층에서 활성화함수(소프트맥스)를 이용해 계산한다.
"""

# 이미지가 컬러일 때 : RGB(빨강,초록,파랑) 3가지를 쓰니 3차원 배열로 표현해서 (4,4) -> (4,4,3)으로 만들고 해도 된다.
# 커널의 크기 또한 (3,3,3)으로 두고 27개의 원소에 27개의 가중치를 곱하고 절편을 더한다.->출력값은 단 하나다. -> (4,4,3) 입력과 (3,3,3) 크기 커널의 합성곱 결과는 (2,2)크기의 특성맵이된다.

# 사실 케라스는 항상 3차원을 기대한다. 만약 2차원의 흑백이라면 (28,28) -> (28,28,1)처럼 3차원 배열로 변환한다.
```
